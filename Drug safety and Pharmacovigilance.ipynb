```python
def Agecheck(age):           #function based column creation
    if(age<35):
        return 'Young'
    else:
        return 'Old'



df['Person Existence']=df['Age'].apply(Agecheck)
df.head()
```

```python
def healthcheck(row):
    if((row['Cholesterol']=='HIGH') and (row['Drug']=='drugC')):
        return 'Bad'
    else:
        return 'Good'

df['Health']=df.apply(healthcheck,axis=1)
df.head()
```

```python
def Sexcheck(sex):           #function based column creation
    if(sex=='M'):
        return 1
    else:
        return 0



df['Sex']=df['Sex'].apply(Sexcheck)
df.head()
```

```python
def CHOcheck(cho):           #function based column creation
    if(cho=='NORMAL'):
        return 1
    else:
        return -1



df['Cholesterol']=df['Cholesterol'].apply(CHOcheck)
df.head()
```

```python
def bpcheck(bp):           #function based column creation
    if(bp=='NORMAL'):
        return 0
    elif bp=='LOW':
        return -1
    else:
        return 1



df['BP']=df['BP'].apply(bpcheck)
df.head()
```

```python
df=df.drop('Health',axis=1)
df=df.drop('Person Existence',axis=1)
df.head()
```

```python
df.groupby('Drug').count()
```

```python
def drugcheck(drug):           #function based column creation
    if(drug=='DrugY'):
        return 1
    elif drug=='drugA':
        return 2
    elif drug=='drugB':
        return 3
    elif drug=='drugC':
        return 4
    else:
        return 5





df['Drug']=df['Drug'].apply(drugcheck)
df.head()
```
